## Data Preprocessing

```python
# Data preparation
user_identifier = dataset['user']
dataset = dataset.drop(columns=['user'])
```

```python
# One hot encoding
dataset.housing.value_counts()
```

```python
dataset.zodiac_sign.value_counts()
```

```python
dataset.payment_type.value_counts()
```

```python
dataset = pd.get_dummies(dataset)
dataset.columns
```

```python
dataset = dataset.drop(columns=['housing_na','zodiac_sign_na','payment_type_na']) # to remove correlation between the categorical data
```

```python
## Splitting the dataset into training set and testing set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(dataset.drop(columns='churn'),
                                                dataset['churn'],
                                                test_size = 0.2,
                                                random_state=0)
```

```python
X_train.head()
```

```python
y_train.value_counts()
```

```python
pos_index = y_train[y_train.values==1].index
neg_index = y_train[y_train.values==0].index
```

```python
if len(pos_index)>len(neg_index):
    higher = pos_index
    lower = neg_index
else:
    lower = pos_index
    higher = neg_index
```

```python
random.seed(0)
higher = np.random.choice(higher,size = len(lower))
lower = np.asarray(lower)
new_indexes = np.concatenate((lower,higher))
```

```python
print(higher.shape)
print(lower.shape)
print(new_indexes.shape)
```

```python
X_train = X_train.loc[new_indexes,]
y_train = y_train[new_indexes]
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
# Feature Scaling

from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train2 = pd.DataFrame(sc_X.fit_transform(X_train)) # StandardScaler returns numpy array and we need the names of columns and rows, so we convert it to dataframe
X_test2 = pd.DataFrame(sc_X.fit_transform(X_test))
X_train2.columns = X_train.columns.values
X_test2.columns = X_test.columns.values
X_train2.index = X_train.index.values
X_test2.index = X_test.index.values
X_train = X_train2
X_test = X_test2
```

## Model Building

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

## Evaluating the results

```python
from sklearn.metrics import confusion_matrix,accuracy_score,f1_score,precision_score,recall_score
cm = confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True,fmt='g')
```

```python
accuracy_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```

```python
f1_score(y_test,y_pred)
```

## Applying k-fold cross validation

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier,
                            X=X_train,
                            y=y_train,
                            cv = 10)
```

```python
accuracies
```

```python
accuracies.mean()
```

```python
# Analyzing coefficients
pd.concat([pd.DataFrame(X_train.columns, columns=['features']),
         pd.DataFrame(np.transpose(classifier.coef_),columns=['coefficient'])],
          axis=1) # classifier.coef_ returns a row so we use np.transpose to convert is to the column
```

## Feature Selection

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
rfe = RFE(classifier,20) # We want to have best 20 columns which predict the result in our final answer
rfe = rfe.fit(X_train,y_train)
```

```python
print(rfe.support_)
```

```python
X_train.columns[rfe.support_]
```

```python
rfe.ranking_
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train[X_train.columns[rfe.support_]],y_train)

y_pred = classifier.predict(X_test[X_test.columns[rfe.support_]])

from sklearn.metrics import confusion_matrix,accuracy_score,f1_score,precision_score,recall_score
cm = confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True,fmt='g')
```

```python
accuracy_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```

```python
f1_score(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```

```python
# Analyzing coefficients
pd.concat([pd.DataFrame(X_train.columns[rfe.support_], columns=['features']),
         pd.DataFrame(np.transpose(classifier.coef_),columns=['coefficient'])],
          axis=1)
```

## Formatting the final results

```python
final_results = pd.concat([y_test,user_identifier],axis=1).dropna()
final_results['predictive_churn'] = y_pred
final_results = final_results[['user','churn','predictive_churn']].reset_index(drop=True)
```
